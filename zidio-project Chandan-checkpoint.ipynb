```python
x = data.iloc[:,2: ].values
x
```

--  **df.iloc[:, 2:]** uses the iloc method to select all rows (:) and all columns starting from the third column (index 2) onwards.
**iloc** is used for integer-location based indexing for selection by position. Here, it selects the slice of the DataFrame starting from the third column to the last column.

```python
scale = MinMaxScaler()
features = scale.fit_transform(x)
features
```

-- An instance of MinMaxScaler is created. By default, it scales the data to the range [0, 1].The fit_transform method is called on the scale object with the data x as its argument.
fit_transform first fits the scaler to the data (calculating the minimum and maximum values for each feature) and then transforms the data according to these calculated values.
The result is a NumPy array features where each feature is scaled to the range [0, 1].

```python
inertia = []
for i in range(1,16):
    k_means = KMeans(n_clusters=i)
    k_means = k_means.fit(features) # The fit method was outside the loop
    inertia.append(k_means.inertia_)
    # clustering algorithm used to partition a dataset into K distinct,non-overlapping subsets (clusters).
# It tries to make the intra-cluster data points as similar as possible while keeping the clusters as distinct as possible.
```

```python
# elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')
plt.xlabel('No of clusters'), plt.ylabel('Inertia')

# Kelbow visualizer
plt.subplot(1,2,2)
kmeans = KMeans()
visualize = KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
```

-- Thus we are arrived at the conclusion from the above  Elbow graph the optimum number of Clusters to be "3".

--***Silhouette Score***
Silhouette Score is a metric used to evaluate the quality of clusters created by a clustering algorithm. It measures how similar an object is to its own cluster compared to other clusters. The score ranges from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

```python
silhouette_avg = []
for i in range(2,16):
    # initialise kmeans
    kmeans = KMeans(n_clusters= i)
    cluster_labels = kmeans.fit_predict(features)
    # silhouette score
    silhouette_avg.append(silhouette_score(features,cluster_labels))
```

-- A for loop is used to iterate over a range of cluster numbers from 1 to 15.
**In each iteration:**
A KMeans instance is created with i clusters (n_clusters=i).
The model is fitted to the scaled data (features) using the fit method.
The inertia of the fitted model is calculated and appended to the inertia list. Inertia is the sum of squared distances of samples to their closest cluster center.

```python
#Plotting K values from average silhoutte score
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis For Optimal k')
plt.show()
# silhoutte score is 0.48 for this model
```

```python
model = KMeans(n_clusters=3)
model = model.fit(features)
```

--  A KMeans instance is created with the specified number of clusters (optimal_clusters), which you would determine based on the elbow plot or another method.
  The model is then fitted to the features data. This step is necessary if the model has not been fitted previously.

```python
y_km = model.predict(features)
centers = model.cluster_centers_
```

--model.predict(features) assigns each data point in features to a cluster.
The result is stored in y_km, which is an array of cluster labels. Each element in y_km corresponds to a data point in features and indicates the cluster it belongs to.
model.cluster_centers_ returns the coordinates of the cluster centers

```python
data['Cluster'] = pd.DataFrame(y_km)
data.to_csv("Cluster_data",index=False)
```
